```python
#단순회귀분석
```

```python
import numpy as np

dataSize = 20
xData = []
yData = []

# 자료의 형태 설정
wi = 3.5
bi = - 90

# 랜덤 데이터 생성
for i in range(dataSize):
    xi = np.random.normal(0.0, 1.0)
    yi = wi * xi + bi + np.random.normal(0.0, 1.0)
    xData.append([xi])
    yData.append([yi])
    
# 플롯 작성
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (6, 5))
plt.plot(xData, yData, 'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import tensorflow as tf
```

```python
# 머신러닝 시작
import tensorflow as tf

w = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = w * xData + b

# mse를 최소화하는 비용함수
loss = tf.reduce_mean(tf.square(y - yData))

optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)

# 머신러닝 반복에 따른 회귀식 변화
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (15, 40))

for step in range(10):
    sess.run(train)
    cf = sess.run(loss)
    wt = sess.run(w)
    bt = sess.run(b)
    print("loss:",loss, "cf:",cf, "wt:",wt, "bt:", bt)
    
    subfig = fig.add_subplot(10, 3, step + 1)
    plt.plot(xData, yData, 'ro')
    plt.plot(xData, wt * xData + bt)
    plt.tight_layout()
    subfig.set_title('Step ' 
                     + str(step + 1) + '] loss = ' + str(round(cf, 2)) 
                     + ', y = ' + str(round(wt[0], 2)) + 'x + ' + str(round(bt[0], 2)))
    subfig.set_xlabel('x')
    subfig.set_ylabel('y')
    
plt.show()
```
